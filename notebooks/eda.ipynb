```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
# nltk.download('stopwords')
nltk.download('wordnet')
```

```python
df = pd.read_csv('../data/jokes.csv', index_col=0)
df.head()
```

Let us analyze the most frequently occuring words (excluding common words such as `bar`, `bartender`, `man`, `walk`, etc, along with common english stopwords)

```python
stopwords = nltk.corpus.stopwords.words('english')
soup = (' '.join(df['joke'])).split()

soup = [word.encode('ascii', 'ignore').decode().casefold() for word in soup]

lemmatizer = WordNetLemmatizer()
lem_words = [lemmatizer.lemmatize(word) for word in soup if word.isalpha() and word not in stopwords]

len(lem_words)
```

```python
from collections import Counter

counter = Counter(lem_words)
most_common = counter.most_common()
most_common
```
